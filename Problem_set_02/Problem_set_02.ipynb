### Problem 01: Query a list of CITY and STATE from the STATION table.

```python
%%sql
select CITY,STATE from STATION;
```

### Problem 02: Query a list of CITY names from STATION for cities that have an even ID number. Print the results in any order, but exclude duplicates from the answer.

```python
%%sql
select CITY from STATION
  where ID%2=0
    order by CITY;
```

### Problem 03: Find the difference between the total number of CITY entries in the table and the number of distinct CITY entries in the table.

```python
%%sql
select count(CITY)-count(distinct CITY) as Difference from STATION;
```

### Problem 04: Query the two cities in STATION with the shortest and longest CITY names, as well as their respective lengths (i.e.: number of characters in the name). If there is more than one smallest or largest city, choose the one that comes first when ordered alphabetically.

```python
%%sql
select CITY,length(CITY) as "Number Of Character" from STATION
  order by 2 
    limit 1;
```

```python
%%sql
select CITY,length(CITY) as "Number Of Character" from STATION
  order by 2 desc
    limit 1;
```

### Problem 05:	Query the list of CITY names starting with vowels (i.e., a, e, i, o, or u) from STATION. Your result cannot contain duplicates.

```python
%%sql
select distinct(CITY) from STATION
  where CITY regexp "^[aeiou]";
```

### Problem 06: Query the list of CITY names ending with vowels (a, e, i, o, u) from STATION. Your result cannot contain duplicates.

```python
%%sql
select distinct(CITY) from STATION
  where CITY regexp "[aeiou]$";
```

### Problem 07: Query the list of CITY names from STATION which have vowels (i.e., a, e, i, o, and u) as both their first and last characters. Your result cannot contain duplicates.

```python
%%sql
select distinct(CITY) from STATION
  where CITY regexp "^[aeiou].*[aeiou]$" ;
```

### Problem 08:	Query the list of CITY names from STATION that do not start with vowels. Your result cannot contain duplicates.

```python
%%sql
select distinct(CITY) from STATION
  where CITY regexp "^[^aeiou]";
```

### Problem 09: Query the list of CITY names from STATION that do not end with vowels. Your result cannot contain duplicates.

```python
%%sql
select distinct(CITY) from STATION
  where CITY regexp "[^aeiou]$";
```

### Problem 10:	Query the list of CITY names from STATION that either do not start with vowels or do not end with vowels. Your result cannot contain duplicates.

```python
%%sql
select distinct(CITY) from STATION
  where CITY regexp "^[^aeiou]|[^aeiou]$";         -- "|" is OR operator --
```

### Problem 11:	Query the list of CITY names from STATION that do not start with vowels and do not end with vowels. Your result cannot contain duplicates.

```python
%%sql
select distinct(CITY) from STATION
  where CITY regexp "^[^aeiou].*[^aeiou]$";
```
