## Creating necessary table

```python
%%sql
DROP TABLE IF EXISTS `stops`;
CREATE TABLE `stops` (
  `id` int(11) NOT NULL,
  `name` varchar(30) DEFAULT NULL,
  PRIMARY KEY (`id`)
) ENGINE=MyISAM DEFAULT CHARSET=latin1;
```

```python
%%sql
desc stops
```

```python
%%sql
DROP TABLE IF EXISTS `route`;
CREATE TABLE `route` (
  `num` varchar(5) NOT NULL,
  `company` varchar(3) NOT NULL,
  `pos` int(11) NOT NULL,
  `stop` int(11) DEFAULT NULL,
  PRIMARY KEY (`num`,`company`,`pos`),
  KEY `stop` (`stop`)
) ENGINE=MyISAM DEFAULT CHARSET=latin1;
```

### Problem 3:	Give the id and the name for the stops on the '4' 'LRT' service.

```python
%%sql
SELECT * 
FROM stops JOIN route
ON stops.id=route.stop
WHERE company='LRT' AND num='4'
```

### Problem 4:	Show the number of routes that visit either London Road (149) or Craiglockhart (53) and which are greater than 1. 

```python
%%sql
SELECT company,num,COUNT(stop) AS Number_of_routes
FROM route
WHERE stop=149 OR stop=53
GROUP BY company,num
HAVING Number_of_routes>1
```

### Problem 5:	Execute the self join so that it shows the services from Craiglockhart (53) to London Road (149).

```python
%%sql
SELECT r1.company,r1.num,r1.stop,r2.stop
FROM route r1 JOIN route r2
ON r1.num=r2.num AND r1.company=r2.company
WHERE r1.stop=53 AND r2.stop=149
```

### Problem 6:	Give a list of all the services which connect stops 115 and 137 ('Haymarket' and 'Leith')

```python
%%sql
SELECT DISTINCT(r1.company),r1.num
FROM route r1 JOIN route r2
ON r1.num=r2.num AND r1.company=r2.company
WHERE r1.stop=115 AND r2.stop=137
```

### Problem 7: Give a list of the services which connect the stops 'Craiglockhart' and 'Tollcross'

```python
%%sql
SELECT r1.company,r1.num
FROM route r1 JOIN route r2 ON (r1.num=r2.num AND r1.company=r2.company)
JOIN stops s1 ON r1.stop=s1.id
JOIN stops s2 ON r2.stop=s2.id
WHERE s1.name='Craiglockhart' AND s2.name='Tollcross'
```

### Problem 8:	Give a distinct list of the stops which may be reached from 'Craiglockhart' by taking one bus, including 'Craiglockhart' itself, offered by the LRT company. Include the company and bus no. of the relevant services..

```python
%%sql
SELECT s2.name,r1.company,r1.num
FROM route r1 JOIN route r2 ON (r1.num=r2.num AND r1.company=r2.company)
JOIN stops s1 ON r1.stop=s1.id
JOIN stops s2 ON r2.stop=s2.id
WHERE s1.name='Craiglockhart' AND r1.company='LRT'
```

### Problem 9: Show the services available from Tollcross.
